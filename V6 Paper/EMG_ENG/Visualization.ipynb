```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Times New Roman'

# Subjects
subjects = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5",
            "Subject 6", "Subject 7", "Subject 8", "Subject 9", "Subject 10", "Average"]

# 데이터
EMG_data = [
    [18.07, 21.51],   # Subject 1 (Hunmin)
    [0,0],               # Subject 2 (Brian)
    [27.72, 30.04],               # Subject 3 (Xianyu)
    [24.85, 18.96],   # Subject 4 (Carlson)
    [28.61, 39.98],   # Subject 5 (Harold)
    [20.11, 19.71],               # Subject 6 (Minjeong)
    [24.46, 11.78],   # Subject 7 (Youngchul)
    [25.17, 24.11],   # Subject 8 (Byeongchan)
    [38.68, 36.82],   # Subject 9 (Jongin)
    [35.21, 33.75],   # Subject 10 (Hongjun)
    [0,0]                # Average (나중에 계산)
]

ENG_data = [
    [17.61, 18.51],   # Subject 1
    [0,0],               # Subject 2
    [31.05, 29.62],               # Subject 3
    [27.11, 28.02],   # Subject 4
    [30.15, 29.68],   # Subject 5
    [18.23, 14.45],               # Subject 6
    [18.29, 26.32],   # Subject 7
    [23.59, 30.7],    # Subject 8
    [52.16, 52.77],   # Subject 9
    [39.34, 47.99],   # Subject 10
    [0,0]                # Average
]

# 평균 & 표준편차
subjects_valid, EMG_means, EMG_stds, ENG_means, ENG_stds = [], [], [], [], []

for subj, e1, e2 in zip(subjects[:-1], EMG_data[:-1], ENG_data[:-1]):  # 마지막 "Average" 제외
    if len(e1) > 0 and len(e2) > 0:
        subjects_valid.append(subj)
        EMG_means.append(np.mean(e1))
        EMG_stds.append(np.std(e1))
        ENG_means.append(np.mean(e2))
        ENG_stds.append(np.std(e2))

# 전체 평균/표준편차 추가
subjects_valid.append("Average")
EMG_means.append(np.mean(EMG_means))
ENG_means.append(np.mean(ENG_means))
EMG_stds.append(np.mean(EMG_stds))
ENG_stds.append(np.mean(ENG_stds))

# barplot
x = np.arange(len(subjects_valid))
width = 0.35

fig, ax = plt.subplots(figsize=(14,5), dpi=250)
bars1 = ax.bar(x - width/2, EMG_means, width, yerr=EMG_stds, capsize=5, label='EMG')
bars2 = ax.bar(x + width/2, ENG_means, width, yerr=ENG_stds, capsize=5, label='ENG')

# label & style
ax.set_ylabel('Accuracy (%)', fontsize=15)
ax.set_xticks(x)
ax.set_xticklabels(subjects_valid, fontsize=13, rotation=0)
ax.legend(fontsize=15)

plt.tight_layout()
plt.show()
```
