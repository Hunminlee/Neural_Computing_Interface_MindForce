```python
import scipy.io
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
```

```python
path = 'C:/Users/hml76/Desktop/Fasikl/Mindforce/Exp_2025-07-30/E9AD0E7DCC2B/raw/8/'

input_path = path + 'Data_20250730_171909.mat'
output_path = path + 'Data_20250730_171909_modified.mat'

# .mat 파일 로드
mat = scipy.io.loadmat(input_path)
label = mat['Data_Cls']  # shape: (1, 1729)

# 변경 대상 값
target = 8

# 값 변경: 0은 그대로, 0이 아닌 값은 target으로
mapped_label = np.where(label == 0, 0, target)

# 저장할 dict 구성 (기존 키 유지 가능)
mat['Data_Cls'] = mapped_label

# .mat 파일로 저장
scipy.io.savemat(output_path, mat)

print(f"수정된 .mat 파일이 저장되었습니다: {output_path}")
```
