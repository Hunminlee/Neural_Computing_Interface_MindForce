```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 (평균값 계산)
subjects = ["Subject 1", "Byeongchan", "Hongjun", "Jongin", "Youngchul", "Harold", "Carlson", "Minjeong", "Xianyu", "Brian"]

EMG = [
    np.mean([18.07, 21.51]),   # Hunmin
    np.mean([25.17, 24.11]),   # Byeongchan
    np.mean([35.21, 33.75]),   # Hongjun
    np.mean([38.68, 36.82]),   # Jongin
    np.mean([24.46, 11.78]),   # Youngchul
    np.mean([28.61, 39.98]),   # Harold
    np.mean([24.85, 18.96]),   # Carlson
    np.mean([20.11, 19.71]),                    # Minjeong
    np.mean([27.72, 30.04]),                    # Xianyu
    np.nan                     # Brian
]

ENG = [
    np.mean([15.61, 15.51]),   # Hunmin
    np.mean([23.59, 30.7]),    # Byeongchan
    np.mean([39.34, 47.99]),   # Hongjun
    np.mean([52.16, 52.77]),   # Jongin
    np.mean([18.29, 26.32]),   # Youngchul
    np.mean([30.15, 29.68]),   # Harold
    np.mean([27.11, 28.02]),   # Carlson
    np.mean([18.23, 14.45]),                    # Minjeong
    np.mean([31.05, 29.62]),                    # Xianyu
    np.nan                     # Brian
]

# NaN 제거 (데이터 없는 subject는 표시 안 함)
valid_idx = [i for i, (e1, e2) in enumerate(zip(EMG, ENG)) if not (np.isnan(e1) or np.isnan(e2))]
subjects_valid = [subjects[i] for i in valid_idx]
EMG_valid = [EMG[i] for i in valid_idx]
ENG_valid = [ENG[i] for i in valid_idx]

# barplot
x = np.arange(len(subjects_valid))
width = 0.35

fig, ax = plt.subplots(figsize=(10,6))
bars1 = ax.bar(x - width/2, EMG_valid, width, label='EMG')
bars2 = ax.bar(x + width/2, ENG_valid, width, label='ENG')

# label & title
ax.set_ylabel('Average Value')
ax.set_title('Subject-wise Comparison of EMG vs ENG')
ax.set_xticks(x)
ax.set_xticklabels(subjects_valid, rotation=45, ha="right")
ax.legend()

# 값 표시
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
